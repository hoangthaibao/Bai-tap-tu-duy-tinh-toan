```python
from math import *
def nt(n):
    if n < 2:
        return False
    else:
        for i in range(2, isqrt(n) + 1):
            if n % i == 0:
                return False
    return True
a, b = map(int, input(). split())

for i in range(a, b + 1):
    if nt(i):
        print(i, end = ' ')
```
